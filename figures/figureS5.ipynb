```python
### Experiment2 - llh

list_beta = [0.1, 0.3, 0.5, 0.7, 0.9]
list_phi = [0.1, 0.3, 0.5, 0.7, 0.9]

for network in  ['BA','WS','SBM', 'twitter']:

    scores_matrix = np.zeros((len(list_beta),len(list_phi)))

    for index_beta, beta in enumerate(list_beta):
        for index_phi, phi in enumerate(list_phi):

            list_scores = []
            list_mat = []

            for it in range(10):

                with open(path_results+'/_llh_experiment2_confusion_matrix_'+str(network)+'_'+str(beta*10)+'_phi_'+str(phi*10)+'_it_'+str(it)+'_premix_different_parameter.pickle', 'rb') as handle:
                    conf = pickle.load(handle)

                score = (conf[0,0] + conf[1,1] + conf[2,2]) / np.sum(conf)
                list_scores.append(score)
                list_mat.append(conf)

            score_local = np.mean(list_scores)
            scores_matrix[index_beta, index_phi] = score_local

            if ((beta == 0.1)&(phi == 0.1)):
                conf_llh_param_known = np.mean(list_mat,axis=0)

    scores_matrix_llh_param_known = scores_matrix
    scores_matrix_llh_param_known = scores_matrix_llh_param_known[::-1]

    conf_llh_param_known = conf_llh_param_known[::-1]
    
    exec('scores_matrix_llh_param_known_'+network+' = scores_matrix_llh_param_known')
```
