```python
### Experiment1 - likelihood

list_beta = [0.1, 0.3, 0.5, 0.7, 0.9]
list_phi = [0.1, 0.3, 0.5, 0.7, 0.9]
network = 'ER'

scores_matrix = np.zeros((len(list_beta),len(list_phi)))

for index_beta, beta in enumerate(list_beta):
    for index_phi, phi in enumerate(list_phi):

        with open(path_results+'/_llh_experiment1_confusion_matrix_'+str(network)+'_star_without_sp_'+str(beta*10)+'_phi_'+str(phi*10)+'.pickle', 'rb') as handle:
            conf = pickle.load(handle)
            
        scores_matrix[index_beta, index_phi] = (conf[0,0] + conf[1,1]) / np.sum(conf)
        
scores_matrix_llh_no_sp = scores_matrix
scores_matrix_llh_no_sp = scores_matrix_llh_no_sp[::-1]
```

### Experiment 2

```python
### Experiment2 - likelihood

list_beta = [0.1, 0.3, 0.5, 0.7, 0.9]
list_phi = [0.1, 0.3, 0.5, 0.7, 0.9]

scores_matrix = np.zeros((len(list_beta),len(list_phi)))

for index_beta, beta in enumerate(list_beta):
    for index_phi, phi in enumerate(list_phi):
        
        list_scores = []
        list_mat = []
        
        for it in range(10):

            with open(path_results+'/_llh_experiment2_confusion_matrix_'+str(network)+'_'+str(beta*10)+'_phi_'+str(phi*10)+'_it_'+str(it)+'_premix_different_parameter.pickle', 'rb') as handle:
                conf = pickle.load(handle)
                
            score = (conf[0,0] + conf[1,1] + conf[2,2]) / np.sum(conf)
            list_scores.append(score)
            list_mat.append(conf)
            
        score_local = np.mean(list_scores)
        scores_matrix[index_beta, index_phi] = score_local
        
        if ((beta == 0.1)&(phi == 0.1)):
            conf_llh_param_known = np.mean(list_mat,axis=0)
        
scores_matrix_llh_param_known = scores_matrix
scores_matrix_llh_param_known = scores_matrix_llh_param_known[::-1]

conf_llh_param_known = conf_llh_param_known[::-1]
```

```python
### Experiment2 - ml

network = 'ER'
name_model = 'random_forest'
name_cascade = 'general'
sample_size = 8000

with open(path_results+'/_ml_experiment2_confusion_matrices_'+network+'_'+name_model+'_'+str(sample_size)+'_premix_different_parameter.pickle', 'rb') as handle:
    conf_ml = pickle.load(handle)
    
# ml_confusion_matrices_list_'+network+'_'+name_model+'_'+name_cascade+'_'+str(sample_size)+'_premix_different_parameter.pickle

with open(path_results+'/_ml_experiment2_scores_matrix_'+network+'_'+name_model+'_'+str(sample_size)+'_premix_different_parameter.pickle', 'rb') as handle:
    scores_matrix_list = pickle.load(handle)
    
conf_ml = np.mean(conf_ml,axis=0)
scores_matrix_ml_param_known = np.mean(scores_matrix_list, axis = 0)
scores_matrix_ml_param_known = scores_matrix_ml_param_known[::-1]

# confusion matrice
conf_ml_param_known = conf_ml[0,0].T[::-1]
```

```python
def highlight_cell(x,y, ax=None, **kwargs):
    rect = plt.Rectangle((x-.5 + 0.05, y-.5 - 0.02), 0.95,0.95, fill=False, **kwargs)
    ax = ax or plt.gca()
    ax.add_patch(rect)
    return rect
```
