# C-Plot the distribution of each of the features

```python
name_features = ['degree',
                 'nber_infected_neighbors',
                 'prop_infected_neighbors', 
                 'sum_stimuli', 
                 'std_stimuli',
                 'nber_stimuli_by_neighbors', 
                 'time_first_infected_neighbor',
                 'time_last_infected_neighbor']

official_name_feature = ['Degree',
                        'Number of infected neighbors',
                        'Proportion of infected neighbors',
                        'Sum of stimuli  ',
                        'Standard deviation of stimuli',
                        'Number of stimuli by neighbor',
                        'Time since the first infected neighbor',
                        'Time since the last infected neighbor']

color_si = '#5e3c99'
color_cp = '#e66101'
color_sp = 'lightseagreen'
color_label = 'midnightblue'
```

```python
# function to smooth the signals when needed

def smoothTriangle(data, degree):
    triangle=np.concatenate((np.arange(degree + 1), np.arange(degree)[::-1])) # up then down
    smoothed=list(data[:degree])

    for i in range(degree, len(data) - degree * 2):
        point=data[i - len(triangle) // 2 : i - len(triangle) // 2 + len(triangle)] * triangle
        smoothed.append(np.sum(point)/np.sum(triangle))
    
    while len(smoothed) < len(data):
        point = data[-(len(data)-len(smoothed)):]
        smoothed.append(np.sum(point)/np.sum(triangle))
        
    return smoothed
```

```python
feature = 'degree'
index_feature = name_features.index(feature)

list_beta = [0.1, 0.3, 0.5, 0.7, 0.9]
list_phi = [0.1, 0.3, 0.5, 0.7, 0.9]
nber_bins = 15
f=30

fig, ax = plt.subplots(len(list_beta),len(list_phi),sharex = True, sharey = True, facecolor='white',gridspec_kw={'wspace':0.2,'hspace':0.2}, figsize = (8, 8))

fig.suptitle(official_name_feature[index_feature], fontsize = f)

for index_beta, beta in enumerate(list_beta): # list_beta
    for index_phi, phi in enumerate(list_phi): # list_phi
        
        x_row = len(list_beta) - index_beta - 1
        y_col = index_phi
        
        exec('df_SI = df_SI_'+str(index_beta))
        exec('df_CP = df_CP_'+str(index_phi))
        
        values_si = list(df_SI[df_SI['contagion']==0][feature])
        values_cp = list(df_CP[df_CP['contagion']==1][feature])
        values_sp = list(df_sp[feature])
        
        y, bins = np.histogram(values_si, bins=[k for k in range(12)])
        y = [k/sum(y) for k in y]
        bin_centers = 0.5*(bins[1:]+bins[:-1])
        ax[x_row, y_col].plot(bin_centers, y, color = color_si, label = 'Sm')
        
        y, bins = np.histogram(values_cp, bins=[k for k in range(12)])
        y = [k/sum(y) for k in y]
        bin_centers = 0.5*(bins[1:]+bins[:-1])
        ax[x_row, y_col].plot(bin_centers, y, color = color_cp, label = 'Cx')
        
        y, bins = np.histogram(values_sp, bins=[k for k in range(12)])
        y = [k/sum(y) for k in y]
        bin_centers = 0.5*(bins[1:]+bins[:-1])
        ax[x_row, y_col].plot(bin_centers, y, color = color_sp, label = 'St')
      
        ax[x_row,0].set_ylabel(round(beta,1), fontsize = f*0.7, labelpad=33, c = color_label)
        ax[4,index_phi].set_xlabel(phi, fontsize = f*0.7, labelpad=30, c = color_label)
                
        ax[x_row, y_col].tick_params(axis='both', which='major', labelsize=f*0.8)
        ax[x_row, y_col].ticklabel_format(axis='y', style='sci', scilimits=(0,0))
        ax[x_row, y_col].yaxis.get_offset_text().set_fontsize(f*0.5)
        
        if x_row == 4:
            ax2 = ax[x_row, y_col].twiny()
            ax2.axis('off')
            ax2.set_xlabel(official_name_feature[index_feature],color = 'black', fontsize = f) # , labelpad=33
            
        ax[x_row, y_col].yaxis.get_offset_text().set_fontsize(f*0.5)
            
ax[0,4].legend(fontsize = f*0.7, loc = (1.1, -0.1))
            
fig.text(0.5, 0.04, official_name_feature[index_feature] , ha='center',color = 'black', fontsize = f*0.8)
fig.text(0.03, 0.5, 'Density', va='center', rotation='vertical',color = 'black', fontsize = f*0.8)

fig.text(0.5, -0.05, r'$\phi$', ha='center',color = 'black', fontsize = f, c = color_label)
fig.text(-0.08, 0.5, r'$\beta$', va='center', rotation='vertical',color = 'black', fontsize = f, c = color_label)

plt.savefig('distri_'+feature+'.pdf',facecolor=fig.get_facecolor(),bbox_inches='tight',transparent=True)
```
